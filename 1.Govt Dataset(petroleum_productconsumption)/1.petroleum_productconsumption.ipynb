## DATASET DESCRIPTION


The dataset represents **monthly product consumption data** over multiple years, containing **492 records** with details about different products and their consumption trends. It consists of five columns: **Month, Year, Product, Quantity, and Updated_Date**. The **Month** column indicates the period of consumption, while the **Year** specifies the corresponding year. The **Product** column lists various consumed products such as Petrol, Diesel, LPG, and ATF. The **Quantity** column provides the amount of each product consumed, measured in metric tonnes or another relevant unit. Lastly, the **Updated_Date** column records the date when the data was last updated. This dataset is useful for analyzing consumption trends, identifying seasonal patterns, and making data-driven decisions regarding product demand across different time periods.

```python
import pandas as pd
df = pd.read_csv("petroleum_productconsumption.csv")
```

```python
df['updated_date'] = pd.to_datetime(df['updated_date'])

```

```python
df.columns = ['Month', 'Year', 'Product', 'Quantity', 'Updated_Date']
```

## Data Analysis & Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Month', y='Quantity', hue='Year', marker='o')
plt.xticks(rotation=45)
plt.title("Monthly Product Consumption Over the Years")
plt.show()
```

### Monthly Consumption Trends

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Product', y='Quantity', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Total Consumption of Different Products")
plt.show()
```

### Product-Wise Consumption

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Year', y='Quantity')
plt.title("Yearly Consumption Distribution")
plt.show()
```

### Yearly Consumption Trend

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Features")
plt.show()
```

### Correlation Analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Month', y='Quantity', hue='Year', marker='o', ci=None)
plt.xticks(rotation=45)
plt.title("Monthly Product Consumption Trend")
plt.show()
```

### Monthly Consumption Trend (Line Plot)

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Product', y='Quantity', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Total Consumption per Product")
plt.show()
```

### Total Consumption per Product (Bar Chart)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Year', y='Quantity')
plt.title("Yearly Consumption Distribution")
plt.show()
```

### Yearly Consumption Box Plot

```python
sns.pairplot(df, diag_kind='kde')
plt.show()
```

### Pair Plot for Numerical Features

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Quantity'], bins=20, edgecolor='black')
plt.title("Distribution of Product Consumption")
plt.xlabel("Quantity (000 Metric Tonnes)")
plt.ylabel("Frequency")
plt.show()
```

### Histogram of Product Consumption

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Year', y='Quantity', hue='Product', style='Product')
plt.title("Yearly Product Consumption Scatter Plot")
plt.show()
```

 ### Scatter Plot of Quantity vs Year

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Product')
plt.xticks(rotation=45)
plt.title("Number of Records per Product")
plt.show()
```

### Count Plot of Product Appearances

```python
df_grouped = df.groupby('Product')['Quantity'].sum()
plt.figure(figsize=(8, 8))
plt.pie(df_grouped, labels=df_grouped.index, autopct='%1.1f%%', startangle=140)
plt.title("Product Consumption Share")
plt.show()
```

###  Pie Chart for Product Consumption Share

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Year', y='Quantity')
plt.title("Regression Plot of Yearly Consumption")
plt.show()
```

###  Regression Plot of Year vs Quantity


```python
top_products = df.groupby('Product')['Quantity'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.index, y=top_products.values)
plt.xticks(rotation=45)
plt.title("Top 10 Most Consumed Products")
plt.show()
```

### Bar Plot of Top 10 Most Consumed Products

```python
df_grouped = df.groupby('Product')['Quantity'].sum()
plt.figure(figsize=(8, 8))
plt.pie(df_grouped, labels=df_grouped.index, autopct='%1.1f%%', startangle=140)
plt.title("Total Product Consumption Share")
plt.show()
```

### Pie Chart for Total Product Consumption Share

```python
top_5_products = df.groupby('Product')['Quantity'].sum().nlargest(5)
plt.figure(figsize=(8, 8))
plt.pie(top_5_products, labels=top_5_products.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Most Consumed Products")
plt.show()
```

### Pie Chart for Top 5 Most Consumed Products

```python
product_name = "ATF"
df_filtered = df[df['Product'] == product_name].groupby('Year')['Quantity'].sum()
plt.figure(figsize=(8, 8))
plt.pie(df_filtered, labels=df_filtered.index, autopct='%1.1f%%', startangle=140)
plt.title(f"Yearly Consumption Share for {product_name}")
plt.show()
```
